```python
db = pd.read_csv("cydec.csv", sep=";")
db = db[0:2000]
db
```

```python
def preprocess(s) :
    s.replace("@", " ")
    return s.replace("#", " ")
    
db["tweets"] = db.tweets.apply(preprocess)  
db[0:2000]
```

```python
lexNorm = pd.read_csv("oov.csv", sep="\t")
lexNorm.dropna(inplace=True)
lexNorm.reset_index(drop=True, inplace=True)
lexNorm.columns = ["oov", "type", "iv"]
lexNorm = lexNorm[lexNorm["oov"] != lexNorm["iv"]]
lexNorm
```

```python
lexNormDict = dict(zip(lexNorm['oov'], lexNorm['iv']))
lexNormDict
```

```python
def normalize(s) :
    l = s.strip().split(" ")
    for i in range(0, len(l)) :
        if l[i] in lexNormDict :
            l[i] = lexNormDict[l[i]]
    return " ".join(l)
db["tweets"] = db.tweets.apply(normalize)
db
```

```python
#importing the glove library
from glove import Corpus, Glove
# creating a corpus object
corpus = Corpus() #for global context
#training the corpus to generate the co occurence matrix which is used in GloVe
corpus.fit(db["tweets"], window=5)
#creating a Glove object which will use the matrix created in the above lines to create embeddings
#We can set the learning rate as it uses Gradient Descent and number of components
glove = Glove(no_components=100, learning_rate=0.01)
 
glove.fit(corpus.matrix, epochs=10, no_threads=4, verbose=True)
glove.add_dictionary(corpus.dictionary)
glove.save('glove.model')
```
